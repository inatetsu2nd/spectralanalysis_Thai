###### <span style="color: gray ">Importing Libraries Together / การนำเข้าไลบรารีพร้อมกัน</span>


```python
# glob: Module for file path pattern matching / โมดูลสำหรับจับคู่แพทเทิร์นของเส้นทางไฟล์
import glob

# NumPy: Library for efficient numerical computations / ไลบรารีสำหรับการคำนวณเชิงตัวเลขอย่างมีประสิทธิภาพ
import numpy as np

# Pandas: Library to facilitate data analysis / ไลบรารีสำหรับการวิเคราะห์ข้อมูล
import pandas as pd

# Matplotlib: Library for plotting graphs / ไลบรารีสำหรับการวาดกราฟ
import matplotlib.pyplot as plt

# Scikit-learn: Library for machine learning / ไลบรารีสำหรับการเรียนรู้ของเครื่อง
from sklearn.preprocessing import StandardScaler  # Module for standardization / โมดูลสำหรับการปรับมาตรฐาน

# SciPy: Library for scientific computing / ไลบรารีสำหรับการคำนวณทางวิทยาศาสตร์และเทคโนโลยี
from scipy.signal import savgol_filter  # Function to apply Savitzky-Golay filter / ฟังก์ชันสำหรับใช้ตัวกรอง Savitzky-Golay
```

# <span style="color:RoyalBlue; ">8.1 Bulk Loading of Spectral Data / การอ่านข้อมูลสเปกตรัมแบบกลุ่ม</span>
###### <span style="color: gray ">Code 8.1 Bulk Loading of CSV Files / โค้ด 8.1 การอ่านไฟล์ CSV แบบกลุ่ม</span>


```python
datapath = "./dataChapter08/"  # Directory path of files to read / เส้นทางโฟลเดอร์ของไฟล์ที่จะอ่าน
# Get list of all CSV file paths / ดึงรายชื่อไฟล์ CSV ทั้งหมด
allfiles = glob.glob('{}*.csv'.format(datapath)) 
df_spectra = pd.DataFrame()  # Initialize DataFrame for concatenation / สร้าง DataFrame ว่างสำหรับรวมข้อมูล

# Loop through each CSV file / วนลูปอ่านไฟล์ CSV แต่ละไฟล์
for file in allfiles:
    df = pd.read_csv(file, header=None)  # Read the file / อ่านไฟล์
    # Concatenate 2nd column data horizontally / รวมข้อมูลคอลัมน์ที่ 2 แบบแนวนอน
    df_spectra = pd.concat([df_spectra, df.iloc[:, 1]], axis=1)

df_spectra.columns = range(1, len(allfiles) + 1)  # Set column names based on file count / ตั้งชื่อคอลัมน์ตามจำนวนไฟล์
df_spectra.index = df.iloc[:, 0]  # Set index from the first column of the first file / ตั้งดัชนีจากคอลัมน์แรกของไฟล์แรก
plt.plot(df_spectra, color='black')  # Plot DataFrame as black line plot / วาดกราฟเส้นสีดำจาก DataFrame
df_spectra = df_spectra.T  # Transpose DataFrame / สลับแถวกับคอลัมน์ของ DataFrame
```

# <span style="color:RoyalBlue; ">8.3 Preprocessing Along Rows (Sample Direction) / การเตรียมข้อมูลตามแถว (ทิศทางตัวอย่าง)</span>
## <span style="color: DeepPink; ">8.3.1 Centering and Scaling / การทำให้เป็นศูนย์กลางและการปรับมาตรฐาน</span>
###### <span style="color: gray ">Code 8.2 Displaying Centered and Scaled Spectra / โค้ด 8.2 การแสดงสเปกตรัมที่ผ่านการทำให้เป็นศูนย์กลางและการปรับมาตรฐาน</span>


```python
# Use StandardScaler / ใช้ StandardScaler
# Instantiate StandardScaler for centering only (with_std=False) / สร้างอินสแตนซ์ StandardScaler สำหรับการทำให้เป็นศูนย์กลางเท่านั้น (with_std=False)
center = StandardScaler(with_std=False)  # Centering (with_std=False) / การทำให้เป็นศูนย์กลาง (with_std=False)
scaler = StandardScaler()  # Scaling / การปรับมาตรฐาน

# Perform fit and transform simultaneously / ทำการฟิตและแปลงข้อมูลพร้อมกัน
df_spectra_centered = center.fit_transform(df_spectra)
df_spectra_scaled = scaler.fit_transform(df_spectra)

# Convert results to DataFrame / แปลงผลลัพธ์เป็น DataFrame
df_spectra_centered = pd.DataFrame(df_spectra_centered, columns=df_spectra.columns)
df_spectra_scaled = pd.DataFrame(df_spectra_scaled, columns=df_spectra.columns)

# Calculate mean and std deviation manually using NumPy / คำนวณค่าเฉลี่ยและส่วนเบี่ยงเบนมาตรฐานด้วย NumPy
mean_manual = np.mean(df_spectra, axis=0)
std_manual = np.std(df_spectra, axis=0, ddof=0)
df_spectra_centered_manual = df_spectra - mean_manual
df_spectra_scaled_manual = (df_spectra - mean_manual) / std_manual
```

###### <span style="color: gray ">Code 8.3 Displaying Centered and Scaled Spectra / โค้ด 8.3 การแสดงสเปกตรัมที่ผ่านการทำให้เป็นศูนย์กลางและการปรับมาตรฐาน</span>


```python
# Create subplots / สร้างกราฟย่อย
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Plot mean spectrum (position: 1,1) / แสดงกราฟสเปกตรัมค่าเฉลี่ย (ตำแหน่ง 1,1)
axes[0, 0].plot(mean_manual, color='magenta')
axes[0, 0].set_title('Mean Spectrum')

# Plot standard deviation spectrum (position: 2,1) / แสดงกราฟสเปกตรัมส่วนเบี่ยงเบนมาตรฐาน (ตำแหน่ง 2,1)
axes[1, 0].plot(std_manual, color='cyan')
axes[1, 0].set_title('Standard Deviation Spectrum')

# Plot centered spectrum calculated by NumPy (position: 1,2) / แสดงกราฟสเปกตรัมที่ทำให้เป็นศูนย์กลางโดยใช้ NumPy (ตำแหน่ง 1,2)
axes[0, 1].plot(df_spectra_centered_manual.T, color='magenta')
axes[0, 1].set_title('Centered Spectrum (NumPy)')

# Plot scaled spectrum calculated by NumPy (position: 2,2) / แสดงกราฟสเปกตรัมที่ปรับมาตรฐานโดยใช้ NumPy (ตำแหน่ง 2,2)
axes[1, 1].plot(df_spectra_scaled_manual.T, color='cyan')
axes[1, 1].set_title('Scaled Spectrum (NumPy)')

# Plot centered spectrum using StandardScaler (position: 1,3) / แสดงกราฟสเปกตรัมที่ทำให้เป็นศูนย์กลางโดยใช้ StandardScaler (ตำแหน่ง 1,3)
axes[0, 2].plot(df_spectra_centered.T, color='magenta')
axes[0, 2].set_title('Centered Spectrum (StandardScaler)')

# Plot scaled spectrum using StandardScaler (position: 2,3) / แสดงกราฟสเปกตรัมที่ปรับมาตรฐานโดยใช้ StandardScaler (ตำแหน่ง 2,3)
axes[1, 2].plot(df_spectra_scaled.T, color='cyan')
axes[1, 2].set_title('Scaled Spectrum (StandardScaler)')

# Show plot / แสดงกราฟ
plt.tight_layout()
plt.show()
```

## <span style="color: DeepPink; ">8.3.2 Multiplicative Scatter Correction (MSC) / การแก้ไขการกระเจิงแบบคูณ (MSC)</span>
###### <span style="color: gray ">Code 8.4 Creating and Executing MSC Function / โค้ด 8.4 การสร้างและเรียกใช้ฟังก์ชัน MSC</span>


```python
# Calculate mean spectrum of df_spectra / คำนวณสเปกตรัมเฉลี่ยของ df_spectra
mean_spectrum = df_spectra.mean()

# Define function to apply MSC (Multiplicative Scatter Correction) / กำหนดฟังก์ชันสำหรับใช้ MSC (การแก้ไขการกระเจิงแบบคูณ)
def apply_msc(sample, reference):
    # Calculate slope and intercept / คำนวณความชันและจุดตัดแกน
    fit = np.polyfit(reference, sample, 1, full=True)
    slope = fit[0][0]
    intercept = fit[0][1]
    # Correct the spectrum / แก้ไขสเปกตรัม
    return (sample - intercept) / slope

# Apply MSC to each spectrum in df_spectra and save to new DataFrame / ใช้ MSC กับสเปกตรัมแต่ละแถวใน df_spectra และบันทึกใน DataFrame ใหม่
df_spectra_msc = df_spectra.apply(lambda x: apply_msc(x, mean_spectrum), axis=1)

# Code to create specified plots / โค้ดสำหรับสร้างกราฟตามที่กำหนด
fig, axes = plt.subplots(2, 1, figsize=(5, 5))

# Plot original spectra / แสดงกราฟสเปกตรัมต้นฉบับ
axes[0].plot(df_spectra.T, color='black')
axes[0].plot(mean_spectrum.T, color='red')
axes[0].set_title('Original Spectra')

# Plot corrected spectra after MSC / แสดงกราฟสเปกตรัมที่ถูกแก้ไขหลัง MSC
axes[1].plot(df_spectra_msc.T, color='black')
axes[1].set_title('Corrected Spectra (MSC Applied)')

plt.tight_layout()
plt.show()
```

# <span style="color:RoyalBlue; ">8.4 Preprocessing Along Columns (Wavelength Direction) / การเตรียมข้อมูลตามคอลัมน์ (ทิศทางความยาวคลื่น)</span>
## <span style="color: DeepPink; ">8.4.2 Performing Smoothing, First Derivative, and Second Derivative Preprocessing / การทำสม่ำเสมอ การหาอนุพันธ์อันดับหนึ่ง และอันดับสอง</span>
###### <span style="color: gray ">Code 8.5 Executing Smoothing, First Derivative, and Second Derivative / โค้ด 8.5 การทำสม่ำเสมอและอนุพันธ์อันดับหนึ่งกับสอง</span>


```python
window_size = 17  # Smoothing window size / ขนาดหน้าต่างสำหรับการเรียบเนียน
poly_order = 2    # Polynomial order for approximation / ลำดับพหุนามสำหรับการประมาณค่า

# Apply Savitzky-Golay filter for smoothing and derivatives / ใช้ตัวกรอง Savitzky-Golay สำหรับการเรียบเนียนและอนุพันธ์
df_spectra_sg_smooth = savgol_filter(df_spectra, window_size, poly_order, axis=1)
df_spectra_sg_1st = savgol_filter(df_spectra, window_size, poly_order, deriv=1, axis=1)
df_spectra_sg_2nd = savgol_filter(df_spectra, window_size, poly_order, deriv=2, axis=1)

fig, axes = plt.subplots(3, 1, figsize=(5, 10))

axes[0].plot(df_spectra_sg_smooth.T, color='black')
axes[0].set_title('Smoothed Spectra / สเปกตรัมที่ผ่านการเรียบเนียน')

axes[1].plot(df_spectra_sg_1st.T, color='magenta')
axes[1].set_title('1st Derivative Spectra / สเปกตรัมอนุพันธ์อันดับ 1')

axes[2].plot(df_spectra_sg_2nd.T, color='cyan')
axes[2].set_title('2nd Derivative Spectra / สเปกตรัมอนุพันธ์อันดับ 2')

plt.tight_layout()
plt.show()
```

## <span style="color: DeepPink; ">8.4.3 Performing Preprocessing Using Standard Normal Variate (SNV) / การเตรียมข้อมูลโดยใช้ Standard Normal Variate (SNV)</span>
###### <span style="color: gray ">Code 8.6 Preprocessing with SNV / โค้ด 8.6 การเตรียมข้อมูลด้วย SNV</span>


```python
def snv(input_data):
    # Subtract mean and divide by standard deviation for each spectrum / ลบค่าเฉลี่ยและหารด้วยส่วนเบี่ยงเบนมาตรฐานสำหรับแต่ละสเปกตรัม
    return (input_data - input_data.mean()) / input_data.std()

# Apply SNV to each row (each sample) in DataFrame / ใช้ SNV กับแต่ละแถว (แต่ละตัวอย่าง) ใน DataFrame
df_spectra_snv = df_spectra.apply(snv, axis=1)

fig, axes = plt.subplots(2, 1, figsize=(5, 7))

axes[0].plot(df_spectra_snv.T, color='magenta')
axes[0].set_title('SNV Spectra / สเปกตรัม SNV')

axes[1].plot(df_spectra_msc.T, color='cyan')
axes[1].set_title('MSC Spectra / สเปกตรัม MSC')

plt.tight_layout()
plt.show()
```
