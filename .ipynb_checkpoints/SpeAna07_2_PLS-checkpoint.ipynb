```python
# NumPy: Library for efficient numerical computations / ไลบรารีสำหรับการคำนวณเชิงตัวเลขอย่างมีประสิทธิภาพ
import numpy as np

# Matplotlib: Library for plotting graphs / ไลบรารีสำหรับการวาดกราฟ
import matplotlib.pyplot as plt

# Pandas: Library to facilitate data analysis / ไลบรารีสำหรับการวิเคราะห์ข้อมูล
import pandas as pd

# SciPy: Library for scientific computing / ไลบรารีสำหรับการคำนวณทางวิทยาศาสตร์และเทคโนโลยี
from scipy.stats import norm  # Module for functions related to normal distribution / โมดูลสำหรับฟังก์ชันที่เกี่ยวข้องกับการแจกแจงแบบปกติ

# Scikit-learn: Library for machine learning / ไลบรารีสำหรับการเรียนรู้ของเครื่อง
from sklearn.cross_decomposition import PLSRegression  # Module for Partial Least Squares regression model / โมดูลสำหรับโมเดลการถดถอยแบบ Partial Least Squares
from sklearn.metrics import mean_squared_error  # Function to calculate mean squared error / ฟังก์ชันสำหรับคำนวณค่าเฉลี่ยของความคลาดเคลื่อนยกกำลังสอง
```

## <span style="color: DeepPink; ">7.2.3 Checking Differences Between Weight Loadings and Loadings Using Python / การตรวจสอบความแตกต่างระหว่าง Weight Loadings และ Loadings ด้วย Python</span>
###### <span style="color: gray ">Code 7.2 Function to Calculate and Display Spectra with Three Overlapping Normal Distributions / โค้ด 7.2 ฟังก์ชันสำหรับคำนวณและแสดงสเปกตรัมที่มีการซ้อนทับของการแจกแจงแบบปกติสามตัว</span>


```python
# Function to combine three normal distributions / ฟังก์ชันรวมการแจกแจงแบบปกติสามตัว
def make_normal(wave, d_std, d_posi, d_inten, noiselevel):
    """
    Uses absorption spectra data (A) and known concentration data (C) to calculate calibration curves (K) and predicted concentrations (predC),
    and plots the absorption spectra and calibration curves.
    / ใช้ข้อมูลสเปกตรัมการดูดกลืน (A) และข้อมูลความเข้มข้นที่ทราบ (C) เพื่อคำนวณเส้นการสอบเทียบ (K) และความเข้มข้นที่ทำนาย (predC)
    และแสดงกราฟสเปกตรัมการดูดกลืนและเส้นการสอบเทียบ

    Parameters:
    wave (numpy.ndarray): Wavelength data / ข้อมูลความยาวคลื่น
    d_std (numpy.ndarray): Standard deviations of normal distributions / ส่วนเบี่ยงเบนมาตรฐานของการแจกแจงปกติ
    d_posi (numpy.ndarray): Positions (means) of normal distributions / ตำแหน่ง (ค่าเฉลี่ย) ของการแจกแจงปกติ
    d_inten (numpy.ndarray): Intensities for each normal distribution per sample / ความเข้มของแต่ละการแจกแจงในแต่ละตัวอย่าง
    noiselevel (float): Noise level / ระดับสัญญาณรบกวน

    Returns:
    numpy.ndarray: Absorption spectra data for each sample / ข้อมูลสเปกตรัมการดูดกลืนของแต่ละตัวอย่าง
    """
    len_wave = len(wave)  # Length of horizontal axis / ความยาวแกน x
    len_sample = len(d_inten)  # Number of samples / จำนวนตัวอย่าง
    len_nd = len(d_std)  # Number of normal distributions combined / จำนวนการแจกแจงปกติที่รวมกัน
    Xall = np.zeros((len_wave, len_sample))  # Zero matrix (rows: wavelengths, columns: samples) / เมทริกซ์ศูนย์ (แถว: ความยาวคลื่น, คอลัมน์: ตัวอย่าง)

    for i in range(len_sample):  # For each sample / สำหรับแต่ละตัวอย่าง
        for k in range(len_nd):  # For each normal distribution / สำหรับแต่ละการแจกแจงปกติ
            Xall[:, i] += d_inten[i, k] * norm.pdf(wave, d_posi[k], d_std[k])
        # Add noise / เพิ่มสัญญาณรบกวน
        Xall[:, i] += noiselevel * np.max(Xall[:, i]) * (0.5 - np.random.randn(len_wave))

    plt.figure(figsize=(10, 6))
    plt.plot(wave, Xall)
    for k in range(len_nd):
        plt.plot(wave, norm.pdf(wave, d_posi[k], d_std[k]), 'black')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

    Xall = Xall.transpose()  # For spectra, rows are samples and columns are wavelengths / สำหรับสเปกตรัม แถวคือ ตัวอย่าง คอลัมน์คือ ความยาวคลื่น
    return Xall
```

###### <span style="color: gray ">Code 7.9 Calculation of Intensity Distributions to Create Correlated Spectra / โค้ด 7.9 การคำนวณการแจกแจงความเข้มเพื่อสร้างสเปกตรัมที่มีความสัมพันธ์กัน</span>


```python
# Create an array 'wave' with integers from 0 to 99 / สร้างอาร์เรย์ 'wave' ที่มีจำนวนเต็มตั้งแต่ 0 ถึง 99
wave = np.arange(0, 100, 1)

# Define arrays for standard deviations (d_std) and positions (d_posi) / กำหนดอาร์เรย์สำหรับส่วนเบี่ยงเบนมาตรฐาน (d_std) และตำแหน่ง (d_posi)
d_std = np.array([7, 6, 8])
d_posi = np.array([30, 60, 80])

# Generate an array d_inten with 100 rows and 2 columns of random values / สร้างอาร์เรย์ d_inten ที่มี 100 แถว 2 คอลัมน์โดยค่ารันดอม
d_inten = np.random.rand(100, 2)

# Create a third column based on the second column of d_inten / สร้างคอลัมน์ที่สามจากคอลัมน์ที่สองของ d_inten
m_X = np.mean(d_inten[:, 1])
sd_X = np.std(d_inten[:, 1])
rho = 0.95
Z = np.random.randn(100)
third_column = m_X + sd_X * ((d_inten[:, 1] - m_X) / sd_X) * rho + Z * np.sqrt(1 - rho**2)

# Normalize third_column to range from 0 to 1 / ทำการปรับสเกล third_column ให้อยู่ในช่วง 0 ถึง 1
third_column = third_column - min(third_column)  # Shift minimum to 0 / เลื่อนค่าน้อยสุดให้เป็น 0
third_column_scale = max(third_column)          # Get max for scaling / หาค่าสูงสุดเพื่อใช้ในการปรับสเกล
third_column = third_column / third_column_scale

# Add third_column to d_inten / เพิ่มคอลัมน์ที่สามเข้าไปใน d_inten
d_inten = np.column_stack((d_inten, third_column))

# Create scatter plot of columns 2 and 3 of d_inten / สร้างกราฟจุดของคอลัมน์ที่ 2 และ 3 ของ d_inten
plt.scatter(d_inten[:, 1], d_inten[:, 2])
plt.xlabel('Column 2')
plt.ylabel('Column 3')
plt.title('Scatter Plot of Column 2 vs Column 3')
plt.show()
```

###### <span style="color: gray ">Code 7.10 Creation of Correlated Spectra / โค้ด 7.10 การสร้างสเปกตรัมที่มีความสัมพันธ์กัน</span>


```python
A=make_normal(wave,d_std,d_posi,d_inten,0.01)
C=d_inten[:,[1,2]]
row_means=C.mean(axis=0).reshape(1,- 1)
C=C-row_means
row_means = A.mean(axis=0).reshape(1,- 1)
A=A-row_means 
```

###### <span style="color: gray ">Code 7.11 Running PLS2 and Displaying Results / โค้ด 7.11 การรัน PLS2 และแสดงผลลัพธ์</span>


```python
# Initialize lists to store R2 scores and RMSE / เตรียมลิสต์เก็บค่า R2 และ RMSE
r2_scores = []
rmse_scores = []

# Set range of components for PLS regression (from 1 to 4) / กำหนดช่วงจำนวนองค์ประกอบสำหรับการถดถอย PLS (1 ถึง 4)
component_range = range(1, 5) 

# Loop over component numbers, train model, and evaluate performance / ลูปตามจำนวนองค์ประกอบ ฝึกโมเดล และประเมินผล
for n_c in component_range:
    pls = PLSRegression(n_components=n_c, scale=False)  # Create PLS regression model / สร้างโมเดลถดถอย PLS
    pls.fit(A, C)  # Fit model with spectral data A and chemical data C / ฝึกโมเดลด้วยข้อมูลสเปกตรัม A และข้อมูลเคมี C
    C_pred = pls.predict(A)  # Predict C using A / ทำนาย C โดยใช้ A
    r2_score = pls.score(A, C)  # Calculate R2 score / คำนวณคะแนน R2
    rmse_score = np.sqrt(mean_squared_error(C, C_pred))  # Calculate RMSE / คำนวณ RMSE
    r2_scores.append(r2_score)  # Append R2 score / เพิ่มค่า R2 ในลิสต์
    rmse_scores.append(rmse_score)  # Append RMSE / เพิ่มค่า RMSE ในลิสต์

# Create and train final PLS regression model with 4 components / สร้างและฝึกโมเดล PLS สุดท้ายด้วย 4 องค์ประกอบ
pls = PLSRegression(n_components=4, scale=False)
pls.fit(A, C)

# Print number of iterations to convergence / แสดงจำนวนรอบที่โมเดลคอนเวิร์จ
print("Number of iterations for convergence:", pls.n_iter_)

# Prepare subplot grid for visualization (5x5) / เตรียมกริดซับกราฟสำหรับแสดงผล (5x5)
fig, axs = plt.subplots(5, 5, figsize=(20, 20))

# Scatter plot of C / กราฟกระจายของ C
axs[0, 0].scatter(C[:, 0], C[:, 1])
axs[0, 0].set_title('C')

# Plot spectra A / แสดงกราฟสเปกตรัม A
axs[0, 1].plot(A.T, color='black')
axs[0, 1].set_title('Spectra')

# Scatter plot of number of components vs RMSE / กราฟกระจายจำนวนองค์ประกอบกับ RMSE
axs[0, 2].scatter(range(1, 5), rmse_scores)
axs[0, 2].set_title('Relation: PLS Components vs RMSE')

# Scatter plot of number of components vs R2 / กราฟกระจายจำนวนองค์ประกอบกับ R2
axs[0, 3].scatter(range(1, 5), r2_scores)
axs[0, 3].set_title('Relation: PLS Components vs R2')

# Generate plots for each component / สร้างกราฟสำหรับแต่ละองค์ประกอบ
for i in range(4):
    # Scores and loadings for C and A / คะแนนและโหลดดิ้งของ C และ A
    xscore = pls.x_scores_[:, i]
    xloading = pls.x_loadings_[:, i]
    yloading = pls.y_loadings_[:, i]
    yconstructed = np.dot(pls.y_scores_[:, :i+1], pls.y_loadings_[:, :i+1].T)
    y_residual = C - yconstructed
    xconstructed = np.dot(pls.x_scores_[:, :i+1], pls.x_loadings_[:, :i+1].T)
    x_residual = A - xconstructed
    ypred = np.dot(pls.x_scores_[:, :i+1], pls.y_loadings_[:, :i+1].T)

    # Scatter plot of actual and reconstructed C / กราฟกระจายของ C จริงและ C ที่สร้างใหม่
    axs[i+1, 0].scatter(C[:, 0], C[:, 1], color='black')
    axs[i+1, 0].scatter(yconstructed[:, 0], yconstructed[:, 1], color='red')
    axs[i+1, 0].set_title(f'PLS Component {i+1} - C (black) and reconstructed C (red)')

    axs[i+1, 1].scatter(C[:, 0], C[:, 1], color='black')
    axs[i+1, 1].scatter(ypred[:, 0], ypred[:, 1], color='blue')
    axs[i+1, 1].set_title(f'PLS Component {i+1} - C (black) and predicted C (blue)')

    # Plot weights and loadings / แสดงกราฟน้ำหนักและโหลดดิ้ง
    axs[i+1, 2].plot(pls.x_weights_[:, i], label='Weights')
    axs[i+1, 2].plot(pls.x_loadings_[:, i], label='Loadings')
    axs[i+1, 2].legend()
    axs[i+1, 2].set_title(f'PLS Component {i+1} - Loadings and Weights')

    # Plot reconstructed A data / แสดงกราฟข้อมูล A ที่สร้างใหม่
    axs[i+1, 3].plot(A.T, color="black")
    axs[i+1, 3].plot(xconstructed.T, color="red")
    axs[i+1, 3].set_title(f'PLS Component {i+1} - A (black) and reconstructed A (red)')

    # Plot residuals / แสดงกราฟส่วนที่เหลือ
    axs[i+1, 4].plot(x_residual.T, color='black')
    axs[i+1, 4].set_title(f'PLS Component {i+1} - Residual')

# Adjust layout and show plot / ปรับรูปแบบและแสดงกราฟ
plt.tight_layout()
plt.show()
```
