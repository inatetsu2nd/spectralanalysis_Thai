# <span style="color:RoyalBlue; ">6.2 Classical Least Squares (CLS) Method / วิธีการสี่เหลี่ยมจัตุรัสน้อยที่สุดแบบคลาสสิก (CLS)</span>
## <span style="color: DeepPink; ">6.2.4 Implementation of CLS Method / การใช้งานวิธี CLS</span>
###### <span style="color: gray ">Code 6.1 Creating a Spectrum with Three Absorptions / โค้ด 6.1 การสร้างสเปกตรัมที่มีการดูดกลืนแสงสามจุด</span>


```python
# Set wavelength range / กำหนดช่วงความยาวคลื่น
wave = np.arange(400, 801, 1)

# Define absorption spectra based on normal distributions for Anthocyanin 1, Anthocyanin 2, Chlorophyll
# / กำหนดสเปกตรัมการดูดกลืนโดยใช้การแจกแจงปกติสำหรับแอนโทไซยานิน 1, แอนโทไซยานิน 2 และคลอโรฟิลล์
antho1 = norm.pdf(wave, 530, 60)  # Center 530nm, std dev 60nm / ศูนย์กลาง 530 นาโนเมตร, ส่วนเบี่ยงเบนมาตรฐาน 60 นาโนเมตร
antho2 = norm.pdf(wave, 550, 50)  # Center 550nm, std dev 50nm / ศูนย์กลาง 550 นาโนเมตร, ส่วนเบี่ยงเบนมาตรฐาน 50 นาโนเมตร
chlo = norm.pdf(wave, 680, 50)    # Center 680nm, std dev 50nm / ศูนย์กลาง 680 นาโนเมตร, ส่วนเบี่ยงเบนมาตรฐาน 50 นาโนเมตร

# Concentration arrays / ชุดความเข้มข้น
concen = np.array([[1, 2, 3],
                   [1, 3, 1],
                   [1, 1, 3],
                   [3, 1, 5],
                   [2, 5, 3]])

# Calculate total absorbance spectra for each concentration / คำนวณสเปกตรัมการดูดกลืนรวมสำหรับแต่ละความเข้มข้น
data = []
for c in concen:
    total_abs = c[0] * antho1 + c[1] * antho2 + c[2] * chlo
    total_abs = total_abs + (1e-4) * (0.5 - np.random.randn(len(wave)))  # Add small noise / เพิ่มสัญญาณรบกวนเล็กน้อย
    data.append(total_abs)
data = np.array(data)

# Plot absorption spectra / แสดงกราฟสเปกตรัมการดูดกลืน
plt.plot(wave, data.T, color='black')
plt.title('Total Absorbance for Different Concentrations')
plt.xlabel('Wavelength (nm)')
plt.ylabel('Absorbance')
plt.show()
```

###### <span style="color: gray ">Code 6.2 Function to Calculate Predicted Concentrations and Display Spectra Using CLS Method / โค้ด 6.2 ฟังก์ชันสำหรับคำนวณความเข้มข้นที่ทำนายและแสดงสเปกตรัมโดยใช้วิธี CLS</span>


```python
def CLSfunction(A, C):
    """
    Parameters:
    A (numpy.ndarray): Absorption spectra data / ข้อมูลสเปกตรัมการดูดกลืน
    C (numpy.ndarray): Concentration data / ข้อมูลความเข้มข้น

    Returns:
    numpy.ndarray: Calibration curves (K) / เส้นการสอบเทียบ (K)
    numpy.ndarray: Predicted concentrations (predC) / ความเข้มข้นที่ทำนาย
    """
    # Create calibration curves / สร้างเส้นการสอบเทียบ
    CT = C.transpose()
    CTA = np.dot(CT, A)
    invCT = np.linalg.inv(np.dot(CT, C))
    K = np.dot(invCT, CTA)

    # Prediction / การทำนาย
    KT = K.transpose()
    AKT = np.dot(A, KT)
    invKK = np.linalg.inv(np.dot(K, KT))
    predC = np.dot(AKT, invKK)

    # Plotting / การสร้างกราฟ
    fig = plt.figure()
    ax1 = fig.add_subplot(2, 1, 1)  # Absorption spectra plot / กราฟสเปกตรัมการดูดกลืน
    ax1.plot(A.transpose())
    ax1.set_title('Absorption Spectra')
    ax1.set_xlabel('Wavelength')
    ax1.set_ylabel('Absorbance')

    ax2 = fig.add_subplot(2, 1, 2)  # Calibration curves plot / กราฟเส้นการสอบเทียบ
    ax2.plot(K.transpose(), 'black')
    ax2.set_title('Calibration Curves')
    ax2.set_xlabel('Wavelength')
    ax2.set_ylabel('Intensity')

    plt.tight_layout()
    plt.show()

    # Display concentrations and predicted concentrations / แสดงความเข้มข้นและความเข้มข้นที่ทำนาย
    print('Concentrations:')
    print(f'{"="*40}')
    print(C, '\n')
    print('Predicted concentrations:')
    print(f'{"="*40}')
    print(np.round(predC, 2))  # Round predC for display / ปัดเศษ predC เพื่อแสดงผล

    return K, predC
```

###### <span style="color: gray ">Code 6.3 Execution of the function CLSfunction / โค้ด 6.3 การรันฟังก์ชัน CLSfunction</span>


```python
# Run CLSfunction with 3 components / รันฟังก์ชัน CLSfunction ด้วย 3 ส่วนประกอบ
K, predC = CLSfunction(data, concen[:, :])  

# Run CLSfunction with 2 components / รันฟังก์ชัน CLSfunction ด้วย 2 ส่วนประกอบ
K, predC = CLSfunction(data, concen[:, [0, 2]])  

# Run CLSfunction with 1 component / รันฟังก์ชัน CLSfunction ด้วย 1 ส่วนประกอบ
K, predC = CLSfunction(data, concen[:, [1]])  
```

# <span style="color:RoyalBlue; ">6.3 Inverse Least Squares (ILS) Method / วิธีการสี่เหลี่ยมจัตุรัสน้อยที่สุดแบบผกผัน (ILS)</span>
## <span style="color: DeepPink; ">6.3.3 Implementation of ILS Method / การใช้งานวิธี ILS</span>
###### <span style="color: gray ">Code 6.4 Function to Calculate Predicted Concentrations and Display Spectra Using ILS Method / โค้ด 6.4 ฟังก์ชันสำหรับคำนวณความเข้มข้นที่ทำนายและแสดงสเปกตรัมโดยใช้วิธี ILS</span>


```python
def ILSfunction(A, C):
    """
    Parameters:
    A (numpy.ndarray): Absorption spectra data / ข้อมูลสเปกตรัมการดูดกลืน
    C (numpy.ndarray): Concentration data / ข้อมูลความเข้มข้น

    Returns:
    numpy.ndarray: Predicted concentrations (predC) / ความเข้มข้นที่ทำนาย
    """
    # Calculate transpose of A (AT) / คำนวณทรานสโพสของ A (AT)
    AT = A.transpose()
    # Calculate dot product of AT and A (ATA) / คำนวณผลคูณจุดของ AT กับ A (ATA)
    ATA = np.dot(AT, A)
    # Calculate inverse of ATA (invATA) / คำนวณเมทริกซ์ผกผันของ ATA (invATA)
    invATA = np.linalg.inv(ATA)
    # Calculate calibration curve (Pils) by dot product of invATA, AT, and C / คำนวณเส้นการสอบเทียบ (Pils) จากผลคูณจุดของ invATA, AT และ C
    ATC = np.dot(AT, C)
    Pils = np.dot(invATA, ATC)
    # Calculate predicted concentration (predC) by dot product of A and Pils / คำนวณความเข้มข้นที่ทำนาย (predC) จากผลคูณจุดของ A กับ Pils
    predC = np.dot(A, Pils)
    # Print actual and predicted concentration data / แสดงข้อมูลความเข้มข้นจริงและความเข้มข้นที่ทำนาย
    print("Data:\n", C)
    print("Predicted values:\n", predC)
    return predC
```

###### <span style="color: gray ">Code 6.5 Execution of the function ILSfunction / โค้ด 6.5 การรันฟังก์ชัน ILSfunction</span>


```python
# Run ILSfunction with 1 component / รันฟังก์ชัน ILSfunction ด้วย 1 ส่วนประกอบ
predC = ILSfunction(data[:, [0, 130, 150, 280, 400]], concen[:, [0]])
```

```python
# Instantiate the model / สร้างอินสแตนซ์ของโมเดล
model = LinearRegression()

# Train (fit) the model / ฝึก (ฟิต) โมเดล
# A is absorption spectra data, C is concentration data / A คือข้อมูลสเปกตรัมการดูดกลืน, C คือข้อมูลความเข้มข้น
model.fit(data[:, [0, 130, 150, 280, 400]], concen[:, [0]])

# Predict / ทำนาย
predC = model.predict(data[:, [0, 130, 150, 280, 400]])
print(predC)
```
