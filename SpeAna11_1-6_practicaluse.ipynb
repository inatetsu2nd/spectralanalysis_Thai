# <span style="color:RoyalBlue; ">Outlier Detection / โค้ด 11.1 การตรวจจับค่าผิดปกติ</span>

###### <span style="color: gray ">Code 11.1 Outlier Detection / โค้ด 11.1 การตรวจจับค่าผิดปกติ</span>


```python
def remove_outliers_and_plot(spec, prop_data, contamination):
    """
    Parameters:
    spec (pandas.DataFrame): Spectral data
    prop_data (pandas.Series): Target variable
    contamination (float or 'auto'): Proportion of outliers or 'auto'

    Returns:
    None
    """
    # ① Apply PCA for dimensionality reduction
    pca = PCA(n_components=4)  # Get first 4 principal components
    spectra_pca = pca.fit_transform(spec)

    # ② Create and apply Isolation Forest model (spectral data)
    model_spec = IsolationForest(contamination=contamination, random_state=0)
    predict_spec = model_spec.fit_predict(spectra_pca)

    # ③ Create and apply Isolation Forest model (target variable)
    model_prop = IsolationForest(contamination=contamination, random_state=0)
    predict_prop = model_prop.fit_predict(prop_data.values.reshape(-1, 1))

    # ④ Get indices of samples judged as outliers
    outlier_idx_spec = np.where(predict_spec == -1)[0]
    outlier_idx_prop = np.where(predict_prop == -1)[0]
    outlier_idx = np.union1d(outlier_idx_spec, outlier_idx_prop)
    print("Outliers from spectral data:", outlier_idx_spec)
    print("Outliers from target variable:", outlier_idx_prop)
    
    # Remove outliers from spectral data and target variable
    spectra_isolated = spec.iloc[predict_spec == 1, :]
    prop_isolated = prop_data.iloc[predict_prop == 1]

    # Plot (scatter plots)
    plt.figure(figsize=(6, 6))
    plt.subplot(3, 1, 1)
    plt.scatter(range(len(prop_data)), prop_data, color='black', label='Normal')
    plt.scatter(outlier_idx, prop_data.iloc[outlier_idx], color='red', label='Outlier')
    plt.xlabel('Sample Number')
    plt.ylabel('Property')
    plt.legend()

    plt.subplot(3, 1, 2)
    plt.scatter(spectra_pca[:, 0], spectra_pca[:, 1], color='black', label='Normal')
    plt.scatter(spectra_pca[outlier_idx, 0], spectra_pca[outlier_idx, 1], color='red', label='Outlier')
    plt.xlabel('PC1 Score')
    plt.ylabel('PC2 Score')
    plt.legend()

    plt.subplot(3, 1, 3)
    plt.scatter(spectra_pca[:, 2], spectra_pca[:, 3], color='black', label='Normal')
    plt.scatter(spectra_pca[outlier_idx, 2], spectra_pca[outlier_idx, 3], color='red', label='Outlier')
    plt.xlabel('PC3 Score')
    plt.ylabel('PC4 Score')
    plt.legend()

    # Plot (PCA loadings)
    plt.figure(figsize=(8, 4))
    loadings = pca.components_
    for i in range(4):
        plt.subplot(2, 2, i + 1)
        plt.plot(loadings[i, :])
        plt.title(f'PC{i + 1} Loadings')
        plt.xlabel('Wavelength Index')
        plt.ylabel('Loading Value')

    # Plot PCA scores vs target variable, highlighting outliers
    plt.figure(figsize=(8, 4))
    for i in range(4):
        plt.subplot(2, 2, i + 1)
        plt.scatter(spectra_pca[:, i], prop_data, color='black', label='Normal')
        plt.scatter(spectra_pca[outlier_idx, i], prop_data.iloc[outlier_idx], color='red', label='Outlier')
        plt.title(f'PC{i + 1} score vs property')
        plt.xlabel('Score')
        plt.ylabel('Property')

    plt.tight_layout()
    plt.show()

    return spectra_isolated, prop_isolated
```

###### <span style="color: gray ">Code 11.2 Execution of the function remove_outliers_and_plot</span>
###### <span style="color: gray ">โค้ด 11.2 การเรียกใช้งานฟังก์ชัน remove_outliers_and_plot</span>

```python
spectra_isolated,prop_isolated=remove_outliers_and_plot(spectra, prop["mc"], contamination=0.05)
```

# <span style="color:RoyalBlue; ">11.4 Weight Loading and Contribution Ratio / น้ำหนักการถ่วงและอัตราส่วนการมีส่วนร่วม</span>

###### <span style="color: gray ">Code 11.6 Cross-Validation and Contribution Analysis of PLS Regression Model / โค้ด 11.6 การตรวจสอบไขว้และการวิเคราะห์อัตราส่วนการมีส่วนร่วมของโมเดลถดถอย PLS</span>


```python
def pls_cv_check(spec, prop_data, n_folds=5, max_components=10, scaleset=True):
    """
    Parameters:
    spec (pandas.DataFrame): Spectral data (rows: samples, columns: wavelengths) /
                              ข้อมูลสเปกตรัม (แถว: ตัวอย่าง, คอลัมน์: ความยาวคลื่น)
    prop_data (pandas.Series): Target variable / ตัวแปรเป้าหมาย
    n_folds (int): Number of cross-validation folds / จำนวนชุดในการทำ cross-validation
    max_components (int): Maximum number of PLS components to try / จำนวนองค์ประกอบ PLS สูงสุดที่ทดลอง
    scaleset (bool): Whether to standardize the spectral data / กำหนดว่าจะปรับมาตรฐานข้อมูลสเปกตรัมหรือไม่

    Returns:
    dict: Dictionary of results / พจนานุกรมผลลัพธ์
    """
    scores_cv = []
    scores_train = []
    explained_var_spec = []
    explained_var_prop = []
    coefficients = []
    pred = []

    # Standardize spectral data (center only if scaleset is False) /
    # ปรับมาตรฐานข้อมูลสเปกตรัม (หาก scaleset=False จะทำแค่การทำให้เป็นศูนย์กลาง)
    if scaleset:
        spec_normalized = (spec - spec.mean()) / spec.std(ddof=1)
    else:
        spec_normalized = spec - spec.mean()

    # ① Increase number of PLS factors / เพิ่มจำนวนองค์ประกอบของ PLS ทีละขั้น
    for n in range(1, max_components + 1):
        pls = PLSRegression(n_components=n, scale=scaleset)
        # ② Calculate cross-validation R² score / คำนวณค่า R² จาก cross-validation
        score_cv = cross_val_score(pls, spec, prop_data, cv=n_folds, scoring='r2').mean()
        pls.fit(spec, prop_data)
        # ③ Calculate R² using full training data / คำนวณค่า R² ด้วยข้อมูลฝึกทั้งหมด
        score_train = pls.score(spec, prop_data)
        scores_cv.append(score_cv)  # ④ Append scores / เพิ่มคะแนนเข้าไปในรายการ
        scores_train.append(score_train)
        # ⑤ Calculate principal component scores / คำนวณคะแนนองค์ประกอบหลัก
        transformed_spec = pls.transform(spec)
        # ⑥ Explained variance of spectra / สัดส่วนความแปรปรวนที่อธิบายได้ของข้อมูลสเปกตรัม
        explained_var_spec.append(np.var(transformed_spec, axis=0).sum() / np.var(spec_normalized, axis=0).sum())
        # ⑦ Predict values / คำนวณค่าที่ทำนายได้
        predicted = pls.predict(spec)
        # ⑧ Explained variance of target variable / สัดส่วนความแปรปรวนที่อธิบายได้ของตัวแปรเป้าหมาย
        explained_var_prop.append(np.var(predicted, axis=0).sum() / np.sum(np.var(prop_data, axis=0)))
        # ⑨ Save regression coefficients for each component / บันทึกค่าสัมประสิทธิ์การถดถอยในแต่ละองค์ประกอบ
        coefficients.append(pls.coef_)
        # ⑩ Save predicted values for each component / บันทึกค่าที่ทำนายได้ของแต่ละองค์ประกอบ
        pred.append(predicted)

    optimal_n = np.argmax(scores_cv) + 1
    loadings = pls.x_loadings_
    weights = pls.x_weights_
    print("最適な主成分数数は", optimal_n)  # The optimal number of components is / จำนวนองค์ประกอบที่เหมาะสมที่สุดคือ

    # ⑪ Return key results as dictionary / ส่งคืนผลลัพธ์สำคัญเป็นพจนานุกรม
    results = {
        'optimal_n_components': optimal_n,
        'pred_value': pred,
        'r2_scores_train': scores_train,
        'r2_scores_cv': scores_cv,
        'explained_variances_spec': explained_var_spec,
        'explained_variances_prop': explained_var_prop,
        'loadings': loadings,
        'weights': weights,
        'regression_coefficients': coefficients
    }

    # Create plots / สร้างกราฟ
    plt.figure(figsize=(8, 8))

    # Plot normalized spectra / กราฟแสดงสเปกตรัมที่ถูกทำให้เป็นมาตรฐาน
    plt.subplot(3, 2, 1)
    plt.plot(spec.columns, spec_normalized.T, color='k', linewidth=0.5)
    plt.xlabel('Wavelength')
    plt.ylabel('Normalized Absorbance')
    plt.title('Normalized Spectra')

    # Number of components vs R2 scores / ความสัมพันธ์ระหว่างจำนวนองค์ประกอบกับค่า R²
    plt.subplot(3, 2, 3)
    plt.plot(range(1, max_components + 1), scores_train, 'k-o', label='Train')
    plt.plot(range(1, max_components + 1), scores_cv, 'r-o', label='CV')
    plt.xlabel('Number of PLS Components')
    plt.ylabel('R2 Score')
    plt.title('PLS Components vs R2 Score')
    plt.legend()

    # Number of components vs explained variance / ความสัมพันธ์ระหว่างจำนวนองค์ประกอบกับความแปรปรวนที่อธิบายได้
    plt.subplot(3, 2, 5)
    plt.plot(range(1, max_components + 1), explained_var_spec, 'm-o', label='Spectra')
    plt.plot(range(1, max_components + 1), explained_var_prop, 'c-o', label='Property')
    plt.xlabel('Number of PLS Components')
    plt.ylabel('Explained Variance')
    plt.title('PLS Components vs Explained Variance')
    plt.legend()

    # Weightloadings plot (Components 1-3) / กราฟน้ำหนักถ่วง (องค์ประกอบ 1-3)
    plt.subplot(3, 2, 2)
    for i in range(3):
        plt.plot(spec.columns, weights[:, i], label=f'Component {i + 1}')
    plt.xlabel('Wavelength')
    plt.ylabel('Weightloading')
    plt.title('Weightloadings (Components 1-3)')
    plt.legend()

    # Weightloadings plot (Components 4-6) / กราฟน้ำหนักถ่วง (องค์ประกอบ 4-6)
    plt.subplot(3, 2, 4)
    for i in range(3, 6):
        plt.plot(spec.columns, weights[:, i], label=f'Component {i + 1}')
    plt.xlabel('Wavelength')
    plt.ylabel('Weightloading')
    plt.title('Weightloadings (Components 4-6)')
    plt.legend()

    # Weightloadings plot (Components 7-10) / กราฟน้ำหนักถ่วง (องค์ประกอบ 7-10)
    plt.subplot(3, 2, 6)
    for i in range(6, 10):
        plt.plot(spec.columns, weights[:, i], label=f'Component {i + 1}')
    plt.xlabel('Wavelength')
    plt.ylabel('Weightloading')
    plt.title('Weightloadings (Components 7-10)')
    plt.legend()

    plt.tight_layout()
    plt.show()

    return results
```

###### <span style="color: gray ">Code 11.7 Execution of the pls_cv_check Function / โค้ด 11.7 การเรียกใช้ฟังก์ชัน pls_cv_check</span>


```python
X_train, X_test, y_train, y_test = train_test_split(spectra, prop["mc"], test_size=0.2, random_state=2)
result=pls_cv_check(X_train, y_train, n_folds=5, max_components=10, scaleset=True)
```

# <span style="color:RoyalBlue; ">11.6 Model Optimization Using Pipeline / การปรับแต่งโมเดลโดยใช้ Pipeline</span>

###### <span style="color: gray ">Code 11.13 Optimization of PLS Regression Model Using Derivative Spectra / โค้ด 11.13 การปรับแต่งโมเดลการถดถอย PLS โดยใช้สเปกตรัมอนุพันธ์</span>


```python
# ① Compute derivative spectra using Savitzky-Golay filter /
# คำนวณสเปกตรัมอนุพันธ์โดยใช้ตัวกรอง Savitzky-Golay
def differentiate_spectra(spectra, order, window_length):
    return savgol_filter(spectra, window_length=window_length, polyorder=2, deriv=order)

# ② Create derivative spectra /
# สร้างสเปกตรัมอนุพันธ์
X_train_1st_derivative = [differentiate_spectra(X_train, order=1, window_length=i) for i in range(3, 14, 2)]
X_train_2nd_derivative = [differentiate_spectra(X_train, order=2, window_length=i) for i in range(3, 14, 2)]

# ③ Determine optimal number of PLS components and evaluate with cross-validation /
# หาจำนวนองค์ประกอบ PLS ที่เหมาะสมที่สุดและประเมินด้วยการตรวจสอบไขว้
def optimize_pls(X, y, max_components=15):
    pls_params = {'n_components': range(1, max_components + 1)}
    pls_grid = GridSearchCV(PLSRegression(scale=True), param_grid=pls_params, cv=5, scoring='r2')
    pls_grid.fit(X, y)
    return pls_grid.best_estimator_, pls_grid.best_score_

# ④ Optimize PLS model using original and derivative spectra /
# ปรับแต่งโมเดล PLS โดยใช้สเปกตรัมต้นฉบับและอนุพันธ์
models = [optimize_pls(X_train, y_train)] + [optimize_pls(X, y_train) for X in X_train_1st_derivative] + [optimize_pls(X, y_train) for X in X_train_2nd_derivative]

# Display results /
# แสดงผลลัพธ์
results = pd.DataFrame({
    'Model': ['Original'] + [f'1st Derivative (window={i})' for i in range(3, 14, 2)] + [f'2nd Derivative (window={i})' for i in range(3, 14, 2)],
    'Optimal Components': [model[0].n_components for model in models],
    'R² (CV)': [round(model[1], 3) for model in models]
})
print(results)
```

###### <span style="color: gray ">Code 11.14 Change in Derivative Spectra by Number of Smoothing Points / โค้ด 11.14 การเปลี่ยนแปลงของสเปกตรัมอนุพันธ์ตามจำนวนจุดการทำให้เรียบ</span>


```python
fig, axes = plt.subplots(6, 2, figsize=(7,14))
for i, (X_1st, X_2nd) in enumerate(zip(X_train_1st_derivative, X_train_2nd_derivative)):
    axes[i, 0].plot(X_train.columns,X_1st.T, color='blue', alpha=0.5)
    axes[i, 0].set_title(f'1st Derivative (window={3 + 2*i})')
    axes[i, 1].plot(X_train.columns,X_2nd.T, color='red', alpha=0.5)
    axes[i, 1].set_title(f'2nd Derivative (window={3 + 2*i})')
plt.tight_layout()
plt.show()
```

###### <span style="color: gray ">Code 11.15 Definition of Pipeline / โค้ด 11.15 การกำหนด Pipeline</span>


```python
pipeline = Pipeline([
    ('smoothing', FunctionTransformer(savgol_filter, kw_args={'window_length': 5, 'polyorder': 2})),
    ('pca', PCA()),
    ('regression', LinearRegression())
])
```

###### <span style="color: gray ">Code 11.16 Definition of Parameter Grid for Hyperparameter Tuning / โค้ด 11.16 การกำหนดกริดพารามิเตอร์สำหรับปรับแต่งไฮเปอร์พารามิเตอร์</span>


```python
# Define parameter grid / กำหนดกริดของพารามิเตอร์
param_grid = {
    'smoothing__kw_args': [{'window_length': w, 'polyorder': 2} for w in [5, 7, 9]],
    'pca__n_components': range(1, 16),
}

# Create instance of GridSearchCV / สร้างอินสแตนซ์ของ GridSearchCV
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='r2')

# Execute grid search / ดำเนินการค้นหาแบบกริด
grid_search.fit(X_train, y_train)

# Display best parameters / แสดงพารามิเตอร์ที่ดีที่สุด
print("Best parameters:")
print(grid_search.best_params_)

# Display score of best model / แสดงคะแนนของโมเดลที่ดีที่สุด
print("Best cross-validation score: {:.2f}".format(grid_search.best_score_))
```

###### <span style="color: gray ">Code 11.17 Performance Evaluation Using Test Set / โค้ด 11.17 การประเมินประสิทธิภาพโดยใช้ชุดทดสอบ</span>


```python
# Predict test set with optimized model / ทำนายชุดทดสอบด้วยโมเดลที่ปรับแต่งแล้ว
predicted = grid_search.predict(X_test)

# Performance evaluation / การประเมินประสิทธิภาพ
r2_predicted = r2_score(y_test, predicted)
mse_predicted = mean_squared_error(y_test, predicted)
print(f'PLS R²: {pls_r2:.2f}, RMSE: {np.sqrt(mse_predicted):.2f}')
```

###### <span style="color: gray ">Code 11.18 Definition of Pipeline and Hyperparameter Optimization Using Various Regression Algorithms / โค้ด 11.18 การกำหนด Pipeline และการปรับแต่งไฮเปอร์พารามิเตอร์โดยใช้ขั้นตอนวิธีถดถอยต่างๆ</span>


```python
# Define pipelines / กำหนด Pipeline
pipelines = {
    'PCR': Pipeline([
        ('pca', PCA()),
        ('regressor', LinearRegression())
    ]),
    'PLS': Pipeline([
        ('regressor', PLSRegression())
    ]),
    'SVR': Pipeline([
        ('regressor', SVR())
    ]),
    'RandomForest': Pipeline([
        ('regressor', RandomForestRegressor())
    ]),
    'GradientBoosting': Pipeline([
        ('regressor', GradientBoostingRegressor())
    ]),
    'XGB': Pipeline([
        ('regressor', XGBRegressor())
    ])
}

# Set hyperparameters / กำหนดไฮเปอร์พารามิเตอร์
param_grids = {
    'PCR': {'pca__n_components': range(1, 11)},
    'PLS': {'regressor__n_components': range(1, 11)},
    'SVR': {'regressor__C': [0.1, 1, 10], 'regressor__gamma': ['scale', 'auto']},
    'RandomForest': {'regressor__n_estimators': [10, 50, 100], 'regressor__max_depth': [None, 5, 10]},
    'GradientBoosting': {'regressor__n_estimators': [100, 200], 'regressor__learning_rate': [0.01, 0.1]},
    'XGB': {'regressor__n_estimators': [100, 200], 'regressor__learning_rate': [0.01, 0.1]}
}

# Execute GridSearchCV for each pipeline / ดำเนินการ GridSearchCV สำหรับแต่ละ Pipeline
results = {}
for name, pipeline in pipelines.items():
    grid = GridSearchCV(pipeline, param_grids[name], cv=5, scoring='r2')
    grid.fit(X_train, y_train)
    results[name] = {
        'Best Parameters': grid.best_params_,
        'Best Score': grid.best_score_
    }

# Display results / แสดงผลลัพธ์
for name, result in results.items():
    print(f'{name}:')
    print(f'Best Parameters: {result["Best Parameters"]}')
    print(f'Best Score: {result["Best Score"]:.2f}\n')
```
