```python
# NumPy: Library for efficient numerical computations / ไลบรารีสำหรับการคำนวณเชิงตัวเลขอย่างมีประสิทธิภาพ
import numpy as np

# Pandas: Library to facilitate data analysis / ไลบรารีสำหรับการวิเคราะห์ข้อมูล
import pandas as pd

# Matplotlib: Library for plotting graphs / ไลบรารีสำหรับการวาดกราฟ
import matplotlib.pyplot as plt

# Scikit-learn: Library for machine learning / ไลบรารีสำหรับการเรียนรู้ของเครื่อง
from sklearn.decomposition import PCA  # Module for Principal Component Analysis (PCA) / โมดูลสำหรับการวิเคราะห์องค์ประกอบหลัก (PCA)

# SciPy: Library for scientific computing / ไลบรารีสำหรับการคำนวณทางวิทยาศาสตร์และเทคโนโลยี
from scipy.stats import norm  # Module for functions related to normal distribution / โมดูลสำหรับฟังก์ชันที่เกี่ยวข้องกับการแจกแจงแบบปกติ
```

# <span style="color:RoyalBlue; ">7.1 Principal Component Analysis (PCA) / การวิเคราะห์องค์ประกอบหลัก (PCA)</span>
## <span style="color: DeepPink; ">7.1.3 Checking the Properties of PCA Loadings Using Python / การตรวจสอบคุณสมบัติของโหลดดิ้ง PCA ด้วย Python</span>
###### <span style="color: gray ">Code 7.1 Calculation of Centered Data, Principal Component Scores, Loadings, and Contribution Ratios / โค้ด 7.1 การคำนวณข้อมูลที่ถูกศูนย์กลาง คะแนนองค์ประกอบหลัก โหลดดิ้ง และอัตราการอธิบายตัวแปร</span>


```python
# Create DataFrame / สร้าง DataFrame
data = {'Lycopene (mg/100g)': [7, 8, 9, 10, 11],
        'Sugar Content (Brix%)': [6, 6, 7, 9, 9]}
df = pd.DataFrame(data)

# Calculate mean and unbiased variance / คำนวณค่าเฉลี่ยและความแปรปรวนแบบไม่ลำเอียง
df_mean = df.mean()
df_var = df.var()

# Add mean and variance to a new DataFrame / เพิ่มค่าเฉลี่ยและความแปรปรวนลงใน DataFrame ใหม่
df_stats = pd.concat([df, pd.DataFrame([df_mean, df_var], index=['Mean', 'Unbiased Variance'])])

# Center the data / ศูนย์กลางข้อมูล
df_centered = df - df_mean

# Perform PCA / ทำ PCA
pca = PCA(n_components=2)
df_pca = pd.DataFrame(pca.fit_transform(df_centered), columns=['PC1', 'PC2'])

# Add PCA results to DataFrame / เพิ่มผลลัพธ์ PCA ลงใน DataFrame
df_pca_stats = pd.concat([df_pca, pd.DataFrame([df_pca.mean(), df_pca.var()], index=['Mean', 'Unbiased Variance'])])
explained_variance_ratio = pca.explained_variance_ratio_

# Calculate PCA loadings / คำนวณโหลดดิ้ง PCA (อิทธิพลขององค์ประกอบหลัก)
loadings = pca.components_.T
df_loadings = pd.DataFrame(loadings, columns=['PC1 Loading', 'PC2 Loading'], index=['Lycopene (mg/kg)', 'Sugar Content (Brix)'])

# Convert contribution ratio to DataFrame / แปลงอัตราการอธิบายตัวแปรเป็น DataFrame
df_explained_variance_ratio = pd.DataFrame(explained_variance_ratio, 
                                           columns=['Contribution Ratio'], 
                                           index=['PC1', 'PC2'])

print("Data:\n", df_stats, "\n")
print("Centered data:\n", df_centered, "\n")
print("Principal component scores:\n", round(df_pca_stats, 2), "\n")
print("Loadings:\n", round(df_loadings, 2), "\n")
print("Contribution ratio:\n", round(df_explained_variance_ratio, 4))
```

###### <span style="color: gray ">Code 7.2 Function to Calculate and Display Spectra with Three Overlapping Normal Distributions / โค้ด 7.2 ฟังก์ชันสำหรับคำนวณและแสดงสเปกตรัมที่มีการซ้อนทับของการแจกแจงแบบปกติสามตัว</span>


```python
# Function to combine three normal distributions / ฟังก์ชันสำหรับรวมการแจกแจงแบบปกติสามตัว
def make_normal(wave, d_std, d_posi, d_inten, noiselevel):
    """
    Parameters:
    wave (numpy.ndarray): Wavelength data / ข้อมูลความยาวคลื่น
    d_std (numpy.ndarray): Standard deviations of normal distributions / ค่าส่วนเบี่ยงเบนมาตรฐานของการแจกแจงแบบปกติ
    d_posi (numpy.ndarray): Positions (means) of normal distributions / ตำแหน่ง (ค่าเฉลี่ย) ของการแจกแจงแบบปกติ
    d_inten (numpy.ndarray): Intensities for each normal distribution in each sample / ความเข้มของแต่ละการแจกแจงในแต่ละตัวอย่าง
    noiselevel (float): Noise level / ระดับสัญญาณรบกวน

    Returns:
    numpy.ndarray: Absorption spectra data for each sample / ข้อมูลสเปกตรัมการดูดกลืนของแต่ละตัวอย่าง
    """
    len_wave = len(wave)  # Length of horizontal axis / ความยาวแกน x
    len_sample = len(d_inten)  # Number of samples / จำนวนตัวอย่าง
    len_nd = len(d_std)  # Number of normal distributions combined / จำนวนการแจกแจงปกติที่รวมกัน
    Xall = np.zeros((len_wave, len_sample))  # Zero matrix (rows: wavelengths, columns: samples) / เมทริกซ์ศูนย์ (แถว: ความยาวคลื่น, คอลัมน์: ตัวอย่าง)

    for i in range(len_sample):  # For each sample / สำหรับแต่ละตัวอย่าง
        for k in range(len_nd):  # For each normal distribution / สำหรับแต่ละการแจกแจงปกติ
            Xall[:, i] = Xall[:, i] + d_inten[i, k] * norm.pdf(wave, d_posi[k], d_std[k])
        # Add noise / เพิ่มสัญญาณรบกวน
        Xall[:, i] = Xall[:, i] + noiselevel * np.max(Xall[:, i]) * (0.5 - np.random.randn(len_wave))

    plt.figure(figsize=(10, 6))
    plt.plot(wave, Xall)
    for k in range(len_nd):
        plt.plot(wave, norm.pdf(wave, d_posi[k], d_std[k]), 'black')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.show()

    Xall = Xall.transpose()  # For spectra, rows are samples, columns are wavelengths / สำหรับสเปกตรัม แถวคือ ตัวอย่าง คอลัมน์คือ ความยาวคลื่น
    return Xall
```

###### <span style="color: gray ">Code 7.3 Execution of the function make_normal / โค้ด 7.3 การรันฟังก์ชัน make_normal</span>


```python
wave=np.arange(0,100,1)
d_std=np.array([7,6,8])
d_posi=np.array([30,60,80])
d_inten=np.array([[1,2,3],[1,3,1],[1,1,3],[3,1,5],[2,5,3]])
nd_Xall=make_normal(wave,d_std,d_posi,d_inten,0.001)
```

###### <span style="color: gray ">Code 7.4 Calculation of Inner Products Between Loadings / โค้ด 7.4 การคำนวณดอทโปรดักต์ระหว่างโหลดดิ้ง</span>


```python
# Instantiate PCA and set number of components / สร้างอินสแตนซ์ PCA และกำหนดจำนวนองค์ประกอบหลัก
pca = PCA(n_components=5)  # Use 5 principal components / ใช้องค์ประกอบหลัก 5 ตัว
pca.fit(nd_Xall)  # Train PCA (calculate principal component directions) / ฝึก PCA (คำนวณทิศทางองค์ประกอบหลัก)
score = pca.transform(nd_Xall)  # Calculate scores (principal component scores) / คำนวณคะแนน (คะแนนองค์ประกอบหลัก)
loading = pca.components_  # Calculate loadings (contributions of each principal component) / คำนวณโหลดดิ้ง (อิทธิพลของแต่ละองค์ประกอบหลัก)
explainedv = pca.explained_variance_ratio_  # Calculate explained variance ratio of each principal component / คำนวณอัตราการอธิบายตัวแปรของแต่ละองค์ประกอบหลัก

check_orth = loading.dot(loading.T)  # Inner products between loading vectors (check orthogonality) / ผลคูณภายในระหว่างเวกเตอร์โหลดดิ้ง (ตรวจสอบความตั้งฉาก)
print(np.round(check_orth, 1))  # Display orthogonality results rounded to 1 decimal place / แสดงผลลัพธ์ความตั้งฉากโดยปัดเศษทศนิยม 1 ตำแหน่ง
```

## <span style="color: DeepPink; ">7.1.6 Checking the Effects of Centering and Scaling on Loadings Using Python / การตรวจสอบผลกระทบของการศูนย์กลางและการปรับมาตรฐานต่อโหลดดิ้งด้วย Python</span>
###### <span style="color: gray ">Code 7.5 Spectral Preprocessing (Centering and Scaling) / โค้ด 7.5 การเตรียมข้อมูลสเปกตรัม (การศูนย์กลางและการปรับมาตรฐาน)</span>


```python
df_Xall = pd.DataFrame(nd_Xall)
df_Xall_meaned = (df_Xall - df_Xall.mean())  # Centering / การทำให้เป็นศูนย์กลาง
df_Xall_normalized = df_Xall_meaned / df_Xall.std(ddof=1)  # Scaling (standardization) / การปรับมาตรฐาน (สเกลลิ่ง)
```

###### <span style="color: gray ">Code 7.6 Function to Perform PCA Using SVD / โค้ด 7.6 ฟังก์ชันสำหรับทำ PCA โดยใช้ SVD</span>

```python
def PCAbySVD(X):
    """
    Parameters:
    X (numpy array): 2D array with samples and variables / อาร์เรย์ 2 มิติที่มีตัวอย่างและตัวแปร

    Returns:
    tuple: Tuple containing explained variance ratios (svd_explained), scores (svd_score), and loadings (svd_loading) / ทูเพิลที่ประกอบด้วยอัตราการอธิบายตัวแปร (svd_explained), คะแนน (svd_score) และโหลดดิ้ง (svd_loading)
    """
    # Perform singular value decomposition / ทำการแยกค่าคี่พิเศษ
    U, singular_values, VT = np.linalg.svd(X, full_matrices=False)
    
    # Calculate explained variance ratios / คำนวณอัตราการอธิบายตัวแปร
    svd_explained = (singular_values ** 2) / np.sum(singular_values ** 2)
    
    # Calculate scores / คำนวณคะแนน
    svd_score = np.dot(U, np.diag(singular_values))
    
    # Loadings are given by VT / โหลดดิ้งได้จาก VT
    svd_loading = VT

    return svd_explained, svd_score, svd_loading
```

###### <span style="color: gray ">Code 7.7 Comparison of Loadings Before and After Preprocessing / โค้ด 7.7 การเปรียบเทียบโหลดดิ้งก่อนและหลังการเตรียมข้อมูล</span>


```python
def PCAbySVD(X):
    U, singular_values, VT = np.linalg.svd(X, full_matrices=False)
    svd_explained = (singular_values ** 2) / np.sum(singular_values ** 2)
    svd_score = np.dot(U, np.diag(singular_values))
    svd_loading = VT
    return svd_explained, svd_score, svd_loading

# PCA on original spectra / PCA บนสเปกตรัมต้นฉบับ
svd_explained, svd_score, svd_loading = PCAbySVD(df_Xall)
score_original = pd.DataFrame(svd_score)
loading_original = pd.DataFrame(svd_loading)
explained_original = pd.DataFrame(svd_explained)

# PCA on mean-centered spectra / PCA บนสเปกตรัมที่ทำการศูนย์กลางค่าเฉลี่ย
num_comp = 5
pca = PCA(n_components=num_comp)
pca.fit(df_Xall)
feature = pca.transform(df_Xall)
score_meaned = pd.DataFrame(feature, columns=["PC{}".format(x+1) for x in range(num_comp)])
loading_meaned = pd.DataFrame(pca.components_, index=["PC{}".format(x+1) for x in range(num_comp)])
explained_meaned = pd.DataFrame(pca.explained_variance_ratio_, index=["PC{}".format(x+1) for x in range(num_comp)])

# PCA on standardized spectra / PCA บนสเปกตรัมที่ผ่านการปรับมาตรฐาน
pca.fit(df_Xall_normalized)
feature = pca.transform(df_Xall_normalized)
score_normalized = pd.DataFrame(feature, columns=["PC{}".format(x+1) for x in range(num_comp)])
loading_normalized = pd.DataFrame(pca.components_, index=["PC{}".format(x+1) for x in range(num_comp)])
explained_normalized = pd.DataFrame(pca.explained_variance_ratio_, index=["PC{}".format(x+1) for x in range(num_comp)])

# Plotting results / แสดงผลลัพธ์
fig, axes = plt.subplots(num_comp + 2, 3, figsize=(16, 12))

# Original, meaned, normalized spectra
axes[0][0].plot(df_Xall.T)
axes[0][1].plot(df_Xall_meaned.T)  # df_Xall_meaned คือข้อมูลที่ถูกศูนย์กลางค่าเฉลี่ย
axes[0][2].plot(df_Xall_normalized.T)

# Explained variance ratios
axes[1][0].plot(explained_original)
axes[1][1].plot(explained_meaned)
axes[1][2].plot(explained_normalized)

# Loadings for each PC
for i in range(num_comp):
    axes[i + 2][0].plot(loading_original.iloc[i, :])
    axes[i + 2][1].plot(loading_meaned.iloc[i, :])
    axes[i + 2][2].plot(loading_normalized.iloc[i, :])

# Scatter plots comparing d_inten and PCA scores
fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
# d_inten should be defined elsewhere / d_inten ควรถูกกำหนดไว้ก่อนหน้า
ax1.scatter(d_inten[:, 1] - d_inten[:, 2], score_meaned.iloc[:, 0])

ax2 = fig.add_subplot(2, 1, 2)
ax2.scatter(np.sum(d_inten, axis=1), score_meaned.iloc[:, 1])

plt.show()
```

## <span style="color: DeepPink; ">7.1.8 Confirming Spectral Reconstruction Using Scores and Loadings with Python / การยืนยันการสร้างสเปกตรัมใหม่โดยใช้คะแนนและโหลดดิ้งด้วย Python</span>
###### <span style="color: gray ">Code 7.8 Reconstructing Spectra from Scores and Loadings / โค้ด 7.8 การสร้างสเปกตรัมใหม่จากคะแนนและโหลดดิ้ง</span>


```python
num_components = 5

# Perform PCA / ทำ PCA
pca = PCA(n_components=num_components)
pca.fit(df_Xall_meaned)
scores = pca.transform(df_Xall_meaned)
loadings = pca.components_
explained_variance = pca.explained_variance_ratio_

# Create figure / สร้างกราฟ
fig, axs = plt.subplots(6, 3, figsize=(15, 20))

# Plot original spectra / แสดงกราฟสเปกตรัมต้นฉบับ
axs[0, 1].plot(df_Xall_meaned.T)
axs[0, 1].set_title('Original Spectra')

# Plot PCA results / แสดงผลลัพธ์ PCA
for i in range(num_components):
    # Plot loadings / แสดงกราฟโหลดดิ้ง
    axs[i + 1, 0].plot(loadings[i, :])
    axs[i + 1, 0].set_title(f'PCA {i + 1} : Loadings')

    # Calculate reconstructed spectra / คำนวณสเปกตรัมที่สร้างขึ้นใหม่
    reconstructed = np.dot(scores[:, :i + 1], loadings[:i + 1, :])
    # Calculate residual spectra (measured - reconstructed) / คำนวณสเปกตรัมส่วนที่เหลือ (วัดได้ - สร้างขึ้นใหม่)
    residual_spectra = df_Xall_meaned - reconstructed

    # Plot residual spectra / แสดงกราฟสเปกตรัมส่วนที่เหลือ
    axs[i + 1, 2].plot(residual_spectra.T)
    axs[i + 1, 2].set_title(f'PCA {i + 1} : (Measured - Reconstructed)')

    # Plot reconstructed and original spectra together / แสดงกราฟสเปกตรัมที่สร้างขึ้นใหม่และต้นฉบับพร้อมกัน
    axs[i + 1, 1].plot(reconstructed.T, color='red')
    axs[i + 1, 1].plot(df_Xall_meaned.T, color='black')
    axs[i + 1, 1].set_title(
        f'PCA {i + 1}: Black - Measured, Red - Reconstructed\n'
        f'Spectrum Explained Variance: {explained_variance[i] * 100:.2f}%'
    )

plt.tight_layout()
plt.show()
```
